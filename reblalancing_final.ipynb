```python
import pandas as pd
import numpy as np
import os
```

```python
folder_list = os.listdir('.\\뾰로롱 후보들')
for folder in folder_list:
    path =f'.\\뾰로롱 후보들\\{folder}'
    file_list = os.listdir(path)

    str_amount = 100000000
    investment = str_amount
    buy_money = {}
    str_year = 2010
    end_year = 2020 + 1
    
    year_li = []
    profit_per_li = []
    profit_li = []
    investment_li = []
    for i in range(str_year, end_year):
        globals()[f'df{i}'] =  pd.read_csv(f'.\\krx_4월_첫거래일\\{i}.csv', encoding='euckr')    
        globals()[f'buy_num{i}'] = {}
        buy = 0
        profit = 0
        names = pd.read_csv(f'{path}\\{file_list[i-str_year]}').Name
        if i == str_year:
            for name in names:
                df = globals()[f'df{i}']
                if len(df[df.종목명 == f'{name}']) == 0:
                    name = names[0]
                    print('변경')
                closing = int(df[df.종목명 == f'{name}']['종가'])
                investment_each = investment // len(names) 
                rest_each = int(investment_each % closing)
                buy_money[f'{name}'] = int((investment_each // closing) * closing)
                globals()[f'buy_num{i}'][f'{name}'] = int(investment_each // int(closing))
                print(name, f'지금가격 {closing}원 ',  globals()[f'buy_num{str_year}'][f'{name}'] , '개 구매')# '투자금', buy_money[f'{name}'], '원')
                profit = 0
        else:
            names_b = pd.read_csv(f'{path}\\{file_list[i-(str_year+1)]}').Name
            for nameb in names_b:
                df = globals()[f'df{i}']
                dfb = globals()[f'df{i-1}']

                if len(dfb[dfb.종목명 == f'{nameb}']['종가']) == 0:       
                    nameb = names_b[0]
                if len(df[df.종목명 == f'{nameb}']['종가']) == 0:
                    closingn = int(dfb[dfb.종목명 == f'{nameb}']['종가']) 
                else:  
                    closingn = int(df[df.종목명 == f'{nameb}']['종가'])
                closingb = int(dfb[dfb.종목명 == f'{nameb}']['종가'])
                profit_each = (int(globals()[f'buy_num{i-1}'][f'{nameb}'])) * (closingn - closingb)
                profit += profit_each
                print(nameb, f'지금가격 {closingn}원 작년가격 {closingb}원', globals()[f'buy_num{i-1}'][f'{nameb}'], '개 판매', f' 1년 수익: {profit_each}원')
            before_profit = investment
            investment += profit
            print('-'*20)    

            for name in names:
                df = globals()[f'df{i}']
                dfb = globals()[f'df{i-1}']
                if len(df[df.종목명 == f'{name}']['종가']) == 0:
                    name = names[0]
                closing = int(df[df.종목명 == f'{name}']['종가'])
                investment_each = investment // len(names) 
                globals()[f'buy_num{i}'][f'{name}'] = int(investment_each // closing)
                print(name, closing,'원', globals()[f'buy_num{i}'][f'{name}'], '개 구매')# '투자금', globals()[f'buy_num{i}'][f'{name}'] * int(closing), '원')
                buy +=  globals()[f'buy_num{i}'][f'{name}'] * closing


        if investment <= 0:
            break
        if i == str_year:
            profit_rate = 0
        else:
            profit_rate = (investment - before_profit) / before_profit * 100
        profit_per = round(profit_rate, 1)
        
        year_li.append(i)
        profit_li.append(profit)
        profit_per_li.append(profit_per)
        investment_li.append(investment)
        print(f'{i}년 총 자산 : {investment}원, 수익 : {profit}원, 수익률 : {profit_per}%')
        print('='*80)
    end_amount = investment

    print(f'{folder} 투자 금액: {str_amount}원, {(end_year - 1) - str_year}년 후 금액: {end_amount}, 총 수익: {(end_amount-str_amount)}, 수익률: {round(((end_amount-str_amount)/str_amount) * 100, 1)}')  
    print('='*120)
    investment_dict = {}
    investment_dict['년도'] = year_li
    investment_dict['금액'] = investment_li
    investment_dict['수익'] = profit_li
    investment_dict['수익율'] = profit_per_li
    
    df = pd.DataFrame(investment_dict)
    # df.to_csv(f'./공식별 수익/{folder}.csv')
```
