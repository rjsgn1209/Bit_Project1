```python
file_list = os.listdir('./Project_csv')
file_list

year = []
for i in range(2020, 2009, -1):
    year.append(i)
    
    
for y in year:  

    columns = ['Name','ROE', 'ROA', 'PER','PBR','PSR','PCR', 'POR', 'GP/A', '시가총액']
    stock_df = pd.DataFrame(columns = columns)
    
    for i,j in enumerate(file_list):
    
        dataset = pd.read_csv("./Project_csv/{}".format(j), index_col=0)

        y = str(y)
        if y in dataset:
            
            Name = j.split('.')[0]
            ROE = dataset.loc['당기순이익', y] / dataset.loc['자본총계',y]
            ROA = dataset.loc['당기순이익',y] / dataset.loc['자산총계',y]
            PER = dataset.loc['시가총액', y] / dataset.loc['당기순이익',y]
            PBR = dataset.loc['시가총액', y] / dataset.loc['자본총계',y]
            PSR = dataset.loc['시가총액', y] / dataset.loc['매출액',y]
            PCR = dataset.loc['시가총액', y] / dataset.loc['영업활동현금흐름',y]
            POR = dataset.loc['시가총액', y] / dataset.loc['영업이익',y]
            GP_A = dataset.loc['영업이익',y] / dataset.loc['자산총계',y]
            total = dataset.loc['시가총액',y]

            stock_a = [Name, ROE, ROA, PER, PBR, PSR, PCR, POR, GP_A, total]

            stock_df.loc[i] = stock_a

   
    stock_df.to_csv('./last_data/{}.csv'.format(y))
    
    
```
