```python
def magic_formulas(df, args, limit=1):
   
    result = 0
    df = df.sort_values('시가총액')[:round(len(df) * limit)]
    for arg in args:
        var = arg[:-1]
        pm = arg[-1]
        if pm == '+':
            rank = df[var].rank(method='dense', ascending=False)
        elif pm == '-':
            rank = df[var].rank(method='dense')
        result += rank
    return result.sort_values()[:30].index.to_list()    
    
```

```python
formula_li = []
for y in range(2020, 2009, -1):
    df = pd.read_csv("./last_data/{}.csv".format(y))
    magic_formula = magic_formulas(df, ['ROE+', 'PER-'])
    formula_li.append(magic_formula)
    magic_formula1 = magic_formulas(df, ['GP/A+', 'PER-'])
    formula_li.append( magic_formula1)
    magic_formula2 = magic_formulas(df, ['GP/A+', 'PER-'], 0.2)
    formula_li.append(magic_formula2)
    ta_da1 = magic_formulas(df, ['ROE+', 'PCR-', 'PER-'])  # 뽀로롱1
    formula_li.append(ta_da1)
    ta_da2 = magic_formulas(df, ['ROE+', 'ROA+', 'PER-'])  # 뽀로롱2
    formula_li.append(ta_da2)
    ta_da3 = magic_formulas(df, ['ROE+'])  # 뽀로롱3
    formula_li.append(ta_da3)
    ta_da4 = magic_formulas(df, ['ROE+', 'PSR-', 'PER-'])  # 뽀로롱4
    formula_li.append(ta_da4)
    ta_da5 = magic_formulas(df, ['ROE+', 'ROA+', 'PBR-'])
    formula_li.append(ta_da5)
    
    for formula in formula_li:
        formula.to_csv('.formulas/{formula}/{y}.csv'.format(formula, y))
    
    
```
